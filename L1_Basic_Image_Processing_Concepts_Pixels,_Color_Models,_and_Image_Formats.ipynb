Pixels

**Definition:** The smallest unit of a digital image, representing a single point of color.

**Structure**: Each pixel contains color information, often represented by numerical values.

**Image as a Matrix:**A digital image can be thought of as a matrix of pixels, where each element in the matrix corresponds to a pixel.

**Pixel Depth:** The number of bits used to represent the color of a pixel. Higher pixel depth means more colors can be represented.

```python

import cv2

```

```python
# Assuming your image is in 'My Drive/images/my_image.jpg'
image_path = 'images.jfif'
```

```python
# Read the image using OpenCV
img = cv2.imread(image_path)
```

```python
import matplotlib.pyplot as plt
```

```python
# Display the image
plt.imshow(img)
plt.show()
```

**Color Models**

**Purpose:** Describe how colors are represented in digital images.

**Common Color Models:**

**RGB (Red, Green, Blue):** Additive color model used in most displays, where colors are created by combining different intensities of red, green, and blue light.

**HSV (Hue, Saturation, Value):** Color model that represents color based on its hue (color), saturation (intensity), and value (brightness).

**CMYK (Cyan, Magenta, Yellow, Black):** Subtractive color model used in printing, where colors are created by subtracting specific colors from white light.

**Grayscale:** A simplified color model where each pixel has only one value representing intensity (black to white).

**Color Channels:** Some color models use multiple channels to represent color information. For example, RGB has three channels (red, green, blue).

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Display the image
plt.imshow(imgGray)
plt.show()
```

```python
imgHsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Display the image
plt.imshow(imgHsv)
plt.show()
```
